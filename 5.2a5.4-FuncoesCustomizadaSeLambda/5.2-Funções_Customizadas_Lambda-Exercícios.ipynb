- Crie uma função em Python chamada ```soma_digitos``` que recebe um número inteiro como parâmetro e retorna a soma dos seus dígitos. Por exemplo, para o número 123, a função deve retornar 6, $(1 + 2 + 3)$.
    - O desafio aqui é separar os dígitos de um número inteiro usando operações aritméticas <br> <br>

- No programa principal solicite ao usuário que insira um número e utilize a função ```soma_digitos``` para calcular e exibir a soma dos seus dígitos.

````python
## Crie aqui a função soma_digitos()

def soma_digitos(numero):
````

````python
## Escreva aqui o programa principal e execute a célula

def soma_digitos(numero):
    #garante que o número seja positivo
    numero = abs(numero)
    soma = 0
    #itera enquanto o número não for zero
    while numero > 0:
        #obtém o último dígito do número
        digito = numero % 10
        #adiciona o dígito à soma
        soma += digito
        #remove o último dígito do número
        numero //= 10
    return soma

#programa principal
if __name__ == "__main__":
    #solicita um número inteiro ao usuário
    numero = int(input("Digite um número inteiro: "))
    #chama a função soma_digitos e obtém o resultado
    resultado = soma_digitos(numero)
    #imprime o resultado
    print(f"A soma dos dígitos de {numero} é {resultado}.")
````

### Q4.

- Crie a função ```inverteValor()``` que recebe um inteiro de qualquer tamanho e retorna esse valor invertido usando apenas operações aritméticas
- Crie a função ```verificaInverso()``` que recebe o valor original e o valor invertido e retorna verdadeiro se ambos forem igualmente par ou igualmente ímpar. Retorne falso caso contrário.
- No programa principal, peça um valor do usuário e imprima o retorno de ambas as funções.


````python
## Crie aqui as funções inverteValor() e verificaInverso()

def inverteValor(numero):
def verificaInverso(original, invertido):
````

````python
## Escreva aqui o programa principal e execute a célula

def inverteValor(numero):
    #garante que o número seja positivo
    numero = abs(numero)
    invertido = 0
    while numero > 0:
        #obtém o último dígito do número
        digito = numero % 10
        #adiciona o dígito ao número invertido
        invertido = invertido * 10 + digito
        #remove o último dígito do número
        numero //= 10
    return invertido

def verificaInverso(original, invertido):
    #verifica se ambos números são pares ou ímpares
    return (original % 2 == invertido % 2)

#programa principal
if __name__ == "__main__":
    #solicita um número inteiro ao usuário
    numero = int(input("Digite um número inteiro: "))
    #chama a função inverteValor e obtém o número invertido
    numero_invertido = inverteValor(numero)
    #chama a função verificaInverso e obtém o resultado da verificação
    resultado_verificacao = verificaInverso(numero, numero_invertido)
    #imprime o número invertido e o resultado da verificação
    print(f"O número invertido é {numero_invertido}.")
    print(f"Os números original e invertido são igualmente par ou ímpar? {'Sim' if resultado_verificacao else 'Não'}.")
````

````python
## Escreva aqui o programa principal 

import math

def calcula_perimetro_triangulo(lado1, lado2, lado3):
    return lado1 + lado2 + lado3

def calcula_perimetro_circulo(raio):
    return 2 * math.pi * raio

def calcula_perimetro_retangulo(lado1, lado2=0):
    if lado2 == 0:
        #caso seja um quadrado
        return 4 * lado1
    else:
        #caso seja um retângulo
        return 2 * (lado1 + lado2)

def main():
    while True:
        #exibe o menu
        print("\n1 - Calcular perímetro triângulo")
        print("2 - Calcular perímetro círculo")
        print("3 - Calcular perímetro retângulo")
        print("4 - Sair")
        
        #solicita a opção desejada
        opcao = int(input("Opção: "))
        
        if opcao == 1:
            #calcula perímetro do triângulo
            print("Digite os três lados do triângulo:")
            lado1 = int(input("Lado 1: "))
            lado2 = int(input("Lado 2: "))
            lado3 = int(input("Lado 3: "))
            perimetro = calcula_perimetro_triangulo(lado1, lado2, lado3)
            print(f"O perímetro do triângulo é: {perimetro}")
        
        elif opcao == 2:
            #calcula perímetro do círculo
            raio = int(input("Digite o raio do círculo: "))
            perimetro = calcula_perimetro_circulo(raio)
            print(f"O perímetro do círculo é: {perimetro:.2f}")
        
        elif opcao == 3:
            #calcula perímetro do retângulo ou quadrado
            lado1 = int(input("Informe o primeiro lado: "))
            lado2 = int(input("Informe o segundo lado (ou 0 se for um quadrado): "))
            perimetro = calcula_perimetro_retangulo(lado1, lado2)
            print(f"O perímetro do retângulo é: {perimetro}")
        
        elif opcao == 4:
            #encerra o programa
            print("Saindo...")
            break
        
        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    main()
````

### Q6.

Crie uma função lambda para verificar se um número é par ou ímpar. Em seguida, solicite ao usuário um número indefinidos de valores (até que o usuário digite 0). Para cada valor de entrada, informe se é par ou ímpar.

Exemplo de interação:
```
Digite os valores que deseja verificar a paridade (digite 0 para finalizar a entrada de dados):
3
ímpar
8
par
12
par
5
ímpar
7ímpar
```

````python
# Função lambda para verificar se um número é par ou ímpar
par_ou_impar = lambda x: 'par' if x % 2 == 0 else 'ímpar'

def main():
    print("Digite os valores que deseja verificar a paridade (digite 0 para finalizar a entrada de dados):")
    
    while True:
        #solicita um número ao usuário
        numero = int(input())
        
        #verifica se o número é 0 para encerrar o loop
        if numero == 0:
            break
        
        #usa a função lambda para determinar se o número é par ou ímpar e imprime o resultado
        resultado = par_ou_impar(numero)
        print(resultado)

if __name__ == "__main__":
    main()
````

### Q7.

Escreva um programa que pergunte ao usuário qual operação ele deseja: maior ou menor.
Em seguida leia uma quantidade indefinida de valores do usuário, até que o usuário digite o valor zero.
Apresente ao final o maior ou menor dos valores digitados de acordo com a escolha do usuário.

**Sua solução deve incluir pelo menos uma função ```lambda```** 

Exemplo de interação:
```
Opções: (1) maior ou (2) menor?
Opção: 1

Digite os valores de entrada. Digite 0 para finalizar a entrada de valores.
4
12
15
1
0

O maior valor é: 15
```

````python
## Escreva aqui o programa principal 

#função lambda para encontrar o maior valor
maior = lambda valores: max(valores)

#função lambda para encontrar o menor valor
menor = lambda valores: min(valores)

def main():
    #solicita ao usuário qual operação deseja
    try:
        opcao = int(input("Opções: (1) maior ou (2) menor?\nOpção: "))
    except ValueError:
        print("Entrada inválida. Por favor, insira um número inteiro.")
        return

    if opcao not in [1, 2]:
        print("Opção inválida. Escolha 1 para maior ou 2 para menor.")
        return

    valores = []

    print("Digite os valores de entrada. Digite 0 para finalizar a entrada de valores.")

    while True:
        try:
            numero = int(input())
        except ValueError:
            print("Entrada inválida. Por favor, insira um número inteiro.")
            continue
        
        #verifica se o número é 0 para encerrar o loop
        if numero == 0:
            break
        
        #adiciona o número à lista de valores
        valores.append(numero)
    
    #verifica se a lista de valores está vazia
    if not valores:
        print("Nenhum valor foi inserido.")
        return
    
    #calcula o resultado com base na opção escolhida
    if opcao == 1:
        resultado = maior(valores)
        print(f"O maior valor é: {resultado}")
    elif opcao == 2:
        resultado = menor(valores)
        print(f"O menor valor é: {resultado}")

if __name__ == "__main__":
    main()
````
